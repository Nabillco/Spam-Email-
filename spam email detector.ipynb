```python
x_train,x_test,y_train,y_test=train_test_split(df.text,df.spam,test_size=0.25)
```

```python
cv=CountVectorizer()
x_train_count=cv.fit_transform(x_train.values)
```

```python
model=MultinomialNB()
model.fit(x_train_count,y_train)
```
